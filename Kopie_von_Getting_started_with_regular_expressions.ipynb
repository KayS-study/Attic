## Basics on using regex with python
The code below shows you how to match basic regex patterns in a text. You will notice the functions `re.search` for finding the first match of a pattern in the text, `re.findall` for finding all matches, and `re.sub` for substituting a pattern within the text with a new string. More details can be found in the [documentation](https://docs.python.org/3/library/re.html) and in [this example](https://www.w3schools.com/python/python_regex.asp).

```python
# For using regular expressions in python, we include re module:
import re

# Define the text to search within
text = "This is a sample text with some numbers like 123 and some emails like test@example.com."

# Define the regular expression pattern
# This pattern looks for one or more digits
pattern = r'\d+'

# Search for the pattern in the text
match = re.search(pattern, text)

if match:
  print(f"Found a match: {match.group(0)}") # Print the matched string

# Find all occurrences of the pattern in the text
matches = re.findall(pattern, text)
print(f"All matches found: {matches}")

# Replace occurrences of the pattern in the text
# This replaces all numbers with the word "NUMBER"
new_text = re.sub(pattern, "NUMBER", text)
print(f"Text after replacement: {new_text}")

# You can modify the 'text' and 'pattern' variables above to experiment with different regular expressions.
# Here are a few more examples of patterns:
# r'\w+@\w+\.\w+'  # Matches email addresses
# r'[A-Z]+'       # Matches one or more uppercase letters
# r'\b\w{4}\b'    # Matches words with exactly 4 letters
```
